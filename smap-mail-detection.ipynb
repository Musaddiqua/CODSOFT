```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression  
from sklearn.metrics import accuracy_score
```

Importing the Dependencies


```python
raw_mail_data = pd.read_csv("/kaggle/input/sms-spam-collection-dataset/spam.csv", encoding='ISO-8859-1')
```

```python
print(raw_mail_data)
```

```python
#Relace null value with null string
mail_data = raw_mail_data.fillna('')
```

```python
# Removing the unwanted columns
mail_data_cleaned = mail_data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)

# Printing the first five rows from the cleaned DataFrame
mail_data_cleaned.head()
```

```python
# cheking the number of rows and columnsin the dataframe 
mail_data.shape
```

Data collection and preprosessing 

```python
#label spam mail as 0 ; ham mail as 1

mail_data.loc[mail_data['v1'] == "spam","v1",] = 0
mail_data.loc[mail_data['v1'] == "ham","v1",] = 1
```

Label encoding

```python
#separating the data as text and label

x = mail_data ["v2"]

y = mail_data["v1"]
```

```python
print(x)
```

```python
print(y)
```

spam = 0

ham = 1 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Ensure that x_train and x_test are lists or pandas Series of text data
# Ensure y_train and y_test are numpy arrays or pandas Series of labels

# Create the TfidfVectorizer
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

# Transform the text data to feature vectors
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

# Convert y_train and y_test to integer type if they are not already
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

feature extraction

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with the training data
model.fit(x_train_features, y_train)
```

Logistic regression model

```python
# Make predictions on the training data
prediction_on_training_data = model.predict(x_train_features)

# Calculate accuracy on the training data
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
```

```python
print('accuracy on training data :' ,accuracy_on_training_data)
```

```python
# Make predictions on the test data
prediction_on_test_data = model.predict(x_test_features)

# Calculate accuracy on the training data
accuracy_on_test_data = accuracy_score(y_test, prediction_on_test_data)
```

```python
print('accuracy on test data :' ,accuracy_on_test_data)
```

Evaluating the trained model

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your h..."]

# convert text to featur vectors
input_data_features = feature_extraction.transform(input_mail)

#making predictions

prediction = model.predict(input_data_features)
print(prediction)

if (prediction[0] == 1):
    print("ham mail")
else:
    print("spam mail")
```
